## Denormalisation

### 1) Directors < Directors_Genre

```python
A = []
for idd in tqdm(Directors.id.unique()):
    temp = Directors_genres.loc[Directors_genres["director_id"]==idd]
    val = json.loads(temp[["genre","prob"]].to_json(orient='records'))
    B = {}
    if(val):
        B = {"id":int(idd),"first_name":str(Directors[Directors["id"]==idd]["first_name"].iloc[0]),
             "last_name":str(Directors[Directors["id"]==idd]["last_name"].iloc[0]), "genres":val}
    else:
        B = {"id":int(idd),"first_name":str(Directors[Directors["id"]==idd]["first_name"].iloc[0]),
             "last_name":str(Directors[Directors["id"]==idd]["last_name"].iloc[0]), "genres":list()}
    A.append(B)
```

```python
with open('Structure1\DirectorsFusion.json', 'w') as f:
    json.dump(A, f)
```

```python
with open("Structure1\DirectorsFusion.json", "r") as f:
    director2 = json.load(f)
```

### 2) Movies < Movies_Genre

```python
A = Movies_genres.groupby(by="movie_id")["genre"].apply(list)
```

```python
A.head()
```

```python
Movies= Movies.set_index("id")
Movies.head()
```

```python
M = Movies.join(A)
M = M.reset_index()
```

```python
M.head()
```

```python
#M['genre'] = M['genre'].replace(np.nan, [])
M.head()
```

```python
result = M.to_json(orient="records")
parsed = json.loads(result)
```

```python
with open('Structure1\MoviesFusion.json', 'w') as f:
    json.dump(parsed, f)
```

```python
with open("Structure1\MoviesFusion.json", "r") as f:
    movie2 = json.load(f)
```

### 3) Movies < Directors

```python
A = Movies_directors.groupby(by="movie_id")["director_id"].apply(list)
```

```python
for film in tqdm(movie2):
    film["directors"] = []
    if(film["id"] in A):
        for idd in A[film["id"]]:
            rep = next(item for item in director2 if item["id"] == idd)
            film["directors"].append(rep)
```

```python
with open('Structure1\Movies_Directors_Genres.json', 'w') as f:
    json.dump(movie2, f)
```

```python
with open('Structure1\Movies_Directors_Genres.json', 'r') as f:
    movie3 = json.load(f)
```
